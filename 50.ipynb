```python
languages = languages[languages["babel_level"]!="0"]
```

```python
log = log[["contributor_id","logtype"]]
```

```python
log.head()
```

```python
languages = languages.rename(columns={'babel_user': 'contributor_id', 'babel_level': 'level'})
```

```python
df = log.merge(languages,on="contributor_id" )
```

```python
df = df[["logtype","level"]]
```

```python
df["#"] = 1
```

```python
df.head()
```

```python
df = df.groupby(by=["level","logtype"]).count()
```

```python
df = df.reset_index()
```

```python
df["#"].sum()
```

```python
#Decidi multiplicarlo por 100 para que se aprecien algunos datos que con 2 decimales no se podian apreciar.
#No muestra la probabilidad pero muestra el porcentaje
df["#"] = (df["#"]/df["#"].sum())*100
```

```python
df.pivot_table(index='level',columns='logtype',values='#')
```
